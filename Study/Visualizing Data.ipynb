```python
from collections import Counter
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
```

```python
decile = lambda grade: grade // 10* 10
```

```python
histogram = Counter(decile(grade)
                   for grade in grades)
histogram
```

```python
plt.bar(histogram.keys(),
       histogram.values(),
       8)
plt.axis([-5,105,0,5])
plt.xticks([10 * i
           for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('Students')
plt.title("Distribution of Exam Grades")
```

```python
variance     = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error  = [x + y 
                for x, y in zip(variance, bias_squared)]
```

```python
xs = range(len(variance))
```

```python
plt.plot(xs, variance, 'g-', label = 'variance')
plt.plot(xs, bias_squared, 'r-.', label = 'r^2')
plt.plot(xs, total_error, 'b:', label = 'total error')
plt.title('The graph')
plt.legend(loc = 9)
```
